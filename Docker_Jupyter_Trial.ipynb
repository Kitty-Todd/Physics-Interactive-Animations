```python
import ipywidgets as widgets
from IPython.display import display, Video
import os

run_button = widgets.Button(description = "Generate Animation", 
                            button_style = "success", # "success", "info", "warning", "danger" or "" 
                            icon = "play")

output = widgets.Output()

def on_button_clicked(b):
    with output:
        output.clear_output()
        print("Calculating physics... rendering animation...")

        os.system("manim -ql -v WARNING your_script.py YourSceneName")

        display(Video("media/videos/your_script/480p15/YourSceneName.mp4", embed=True))

run_button.on_click(on_button_clicked)

display(widgets.HTML("<h3>Physics Simulation: Projectile Mostion</h3>"))
display(run_button, output)
```
