```python
import ipywidgets as widgets
import json
import os
from Ipython.display import display, Video

g_slider = widgets.FloatSlider(value=9.8, min=1.0, max = 20.0, description="Gravity: ")
v_slider = widgets.FloatSlider(value=5.0, min=1.0, max=15.0, description="Velocity: ")
run_btn = widgets.Button(description="Launch Simulation", button_style="primary")
output = widgets.Output()

def run_sim(b):
    with output:
        output.clear_output()

        data = {"gravity": g_slider.value, "velocity": v_slider.value}
        with open("params.json", "w") as f:
            json.dump(data, f)
        
        print("Rendering you custom physics scene...")

        os.system("manim -ql -v WARNING phys_sim.py ProjectileSim")

        video_path = "media/videos/phys_sim/480p15/ProjectileSim.mp4"
        if os.path.exists(video_path):
            display(Video(video_path, embed=True, width=600))
        else:
            print("Error: Video not found. Check your script names!")

run_btn.on_click(run_sim)

display(widgets.VBox([widgets.HTML("<h2>Physics Lab: Gravity Explorer</h2>"),
                      g_slider,
                      v_slider,
                      run_btn,
                      output]))
```
